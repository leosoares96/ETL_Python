# Carregando dados com Pandas

```python
import pandas as pd
```

#### Extração do arquivo

```python
data = pd.read_csv('./data/historical-senate-predictions.csv', sep=',', encoding='utf-8')
```

#### Salvando em Excel

```python
data.to_excel('./data/historical-senate-predictions_new.xlsx')
```

#### Salvando em CSV

```python
data.to_csv('./data/historical-senate-predictions_new.csv')
```

#### Importar biblioteca SQLAlchemy

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///database.db')
```

#### Salvando no Banco de Dados sqlite

```python
data.to_sql('senate_predictions',con=engine)
```

#### Consultando o Banco de Dados

```python
engine.execute('SELECT * FROM senate_predictions').fetchall()
```

#### Consultando o Banco de Dados com Pandas

```python
pd.read_sql_query('SELECT * FROM senate_predictions',engine)
```
